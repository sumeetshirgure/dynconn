```python
c = QuantumCircuit(5)

c.u(np.pi/2, 0, np.pi, 4)
c.p(np.pi/4, 4)

c.cx(4, 2)
c.p(-np.pi/4, 2)
c.cx(4, 2)
c.p(np.pi/4, 2)
c.p(np.pi/8, 4)

c.cx(4, 0)
c.p(-np.pi/8, 0)
c.cx(4, 0)
c.p(np.pi/8, 0)

c.barrier()

c.u(np.pi/2, 0, np.pi, 2)
c.p(np.pi/4, 2)

c.cx(2, 0)
c.p(-np.pi/4, 0)
c.cx(2, 0)
c.p(np.pi/4, 0)

c.barrier()

c.u(np.pi/2, 0, np.pi, 0)

c.barrier()

c.draw(output='mpl', style={'dpi': 500}).savefig('decomposed_qft_circuit.png')
c.draw(output='mpl', style={'dpi': 500})
```
