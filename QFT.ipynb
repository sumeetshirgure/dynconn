```python
import numpy as np

from bqskit import Circuit, compile
from bqskit.ir.gates import CPGate, CSGate, CTGate
from bqskit.ext import bqskit_to_qiskit

import qiskit
from qiskit import QuantumCircuit, QuantumRegister
```

```python
qft = qiskit.circuit.library.QFT(5, approximation_degree=1, do_swaps=False, insert_barriers=True)
qft = qft.decompose()
qft.draw(output='mpl')
```

```python
circ = Circuit(5)
circ.append_gate(CTGate(), (0, 4))

circ = compile(circ, optimization_level=4)

qc = bqskit_to_qiskit(circ)
```

```python
qc.draw(output='mpl')
```
