```python
import numpy as np

from bqskit import Circuit, compile
from bqskit.ir.gates import CPGate, CSGate, CTGate, SwapGate
from bqskit.ext import bqskit_to_qiskit

import qiskit
from qiskit import QuantumCircuit, QuantumRegister
```

```python
qft = qiskit.circuit.library.QFT(5, approximation_degree=1, do_swaps=False, insert_barriers=True)
qft = qft.decompose()
qft.draw(output='mpl')
```

```python
sw = QuantumCircuit(2)
sw.swap(0, 1)

sw = sw.decompose()
sw.draw(output='mpl')
```

```python
circ = Circuit(2)
circ.append_gate(SwapGate(), (0, 1))

circ = compile(circ, optimization_level=4)

qc = bqskit_to_qiskit(circ)
qc.draw(output='mpl')
```

```python
circ = Circuit(5)
circ.append_gate(CTGate(), (0, 4))

circ = compile(circ, optimization_level=4)

qc = bqskit_to_qiskit(circ)
qc.draw(output='mpl')
```
